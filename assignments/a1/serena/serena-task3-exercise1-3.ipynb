```python
#using python's pendulum library
from datetime import datetime, timedelta
import pendulum
#this function returns the time from now until 5pm the next day
def time_until_5_tmrw(currtime = None):
    if(currtime==None):
        currtime = pendulum.now('local')
    else: currtime = pendulum.instance(currtime) #set to be a pendulum object
    timetmrw = currtime.add(days=1) #add one day
    tmrw5 = timetmrw.replace(hour=17, minute=0, second=0)
    return (tmrw5 - currtime).total_seconds() / 3600
```
