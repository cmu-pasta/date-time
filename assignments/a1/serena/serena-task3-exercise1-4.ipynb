```python
#using python's pendulum library
from datetime import datetime, timedelta
import pendulum
#this function returns the time from now until 5pm the next day
def time_until_5_tmrw(currtime = None):
    if(currtime==None):
        currtime = pendulum.now('local')
    else: 
        currtime = pendulum.instance(currtime) #set to be a pendulum object
        print(currtime.is_dst())
    timetmrw = currtime.add(days=1) #add one day
    tmrw5 = timetmrw.replace(hour=17, minute=0, second=0)
    return (tmrw5 - currtime).total_seconds() / 3600
```

```python
#old version to see the differences
#using python's datetime library (https://docs.python.org/3/library/datetime.html)
from datetime import datetime, timedelta
#this function returns the time from now until 5pm the next day
def time_until_5_tmrw_old(currtime = None):
    timenow = datetime.now()
    if(currtime!=None): timenow = currtime
    timetmrw = timenow + timedelta(days=1) #add one day
    tmrw5 = timetmrw.replace(hour=17, minute=0, second=0)
    return (tmrw5 - timenow).total_seconds() / 3600
```

```python
#using python's datetime library (https://docs.python.org/3/library/datetime.html)
from datetime import datetime, timedelta
#this function returns the time from now until 5pm the next day
def time_until_5_tmrw_no_tz(currtime = None):
    timenow = datetime.now()
    if(currtime!=None): timenow = currtime
    timetmrw = timenow + timedelta(days=1) #add one day
    tmrw5 = datetime(timetmrw.year, timetmrw.month, timetmrw.day, 17, 0, 0)
    return (tmrw5 - timenow).total_seconds() / 3600
```
